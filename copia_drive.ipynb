```
import numpy as np

# Definición de los vectores
u = np.array([4, -3, 6])
v = np.array([19, 0, 1])
w = np.array([250, 6, -27])
z = np.array([7, -2, 158])

# a) (0.2) u · z + w · (3v)
dot_product_a = np.dot(u, z) + np.dot(w, 3 * v)
print(f"a) Resultado: {dot_product_a:.2f}")

# b) (0.2) (u + z - v) · (z - 3w + v)
vec_b1 = u + z - v
vec_b2 = z - 3 * w + v
dot_product_b = np.dot(vec_b1, vec_b2)
print(f"b) Resultado: {dot_product_b:.2f}")

# c) (0.2) ||z - w|| - ||w × u||
norm_c1 = np.linalg.norm(z - w)
cross_product_wu = np.cross(w, u)
norm_c2 = np.linalg.norm(cross_product_wu)
result_c = norm_c1 - norm_c2
print(f"c) Resultado: {result_c:.2f}")

# d) (0.2) ||u||(v + z) + ||2z - 3w||z
norm_u = np.linalg.norm(u)
vec_d1 = v + z
term_d1 = norm_u * vec_d1
norm_d2 = np.linalg.norm(2 * z - 3 * w)
term_d2 = norm_d2 * z
result_d = term_d1 + term_d2
print(f"d) Resultado: {result_d}")
```

```
import numpy as np

# Definición de los vectores u y v
u = np.array([2, 5])
v = np.array([6, -3])

# Calcular el escalar k para que p = k * u
k = np.dot(v, u) / np.dot(u, u)

# Calcular los vectores p y h
p = k * u
h = v - p

print(f"Vector p paralelo a u: {p}")
print(f"Vector h ortogonal a u: {h}")


import matplotlib.pyplot as plt

# Crear la gráfica
fig, ax = plt.subplots()

# Graficar los vectores p, h, y v
ax.quiver(0, 0, p[0], p[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector p (paralelo a u)')
ax.quiver(0, 0, h[0], h[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector h (ortogonal a u)')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='k', label='Vector v')

# Configurar límites del gráfico
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# Mostrar cuadrícula y leyenda
plt.grid()
plt.legend()
plt.title("Gráfico de los vectores p, h y v")
plt.show()


# Definición de la matriz A
A = np.array([[0, -1], [1, 0]])

# Calcular el vector z = A * u
z = A @ u

print(f"Vector z = A * u: {z}")

# Graficar los vectores u y z
fig, ax = plt.subplots()

ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector u')
ax.quiver(0, 0, z[0], z[1], angles='xy', scale_units='xy', scale=1, color='m', label='Vector z')

# Configurar límites del gráfico
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# Mostrar cuadrícula y leyenda
plt.grid()
plt.legend()
plt.title("Gráfico de los vectores u y z")
plt.show()
```

```
import numpy as np

# Definición de las matrices A, B y C
A = np.array([[7, 2, 4, -5, 2],
              [-4, -4, 3, 6, 0],
              [5, 6, 97, 8, 4],
              [13, 0, 30, 3, 3]])

B = np.array([[-2, 5, 0, -15, -1],
              [7, 6, 8, -20, 0],
              [3, 0, 0, 0, 10],
              [2, -5, -2, 10, 9]])

C = np.array([[5, -8, 0, -20, 1],
              [-2, -4, -3, 5, 2],
              [-5, 0, -4, 50, 24],
              [0, 90, -5, -3, 4]])

# (a) Calcular la matriz X para la primera ecuación
X_a = (2 * A - C + B) / 3
print("Matriz X para la ecuación (a):")
print(X_a)

# (b) Calcular la matriz X para la segunda ecuación
X_b = -(B + 2 * C - 6 * A) / 2
print("\nMatriz X para la ecuación (b):")
print(X_b)
```

```
import numpy as np

# Definir la matriz A basada en los valores de temperatura t
A = np.array([
    [6**3, 6**2, 6, 1],
    [9**3, 9**2, 9, 1],
    [12**3, 12**2, 12, 1],
    [15**3, 15**2, 15, 1]
])

# Definir el vector b basado en los valores de vida media y
b = np.array([134, 160.3, 181.1, 210])

# Resolver el sistema de ecuaciones para encontrar los coeficientes a, b, c, d
coefficients = np.linalg.solve(A, b)

a, b, c, d = coefficients
print(f"Coeficientes: a = {a:.4f}, b = {b:.4f}, c = {c:.4f}, d = {d:.4f}")


# Calcular la vida media a 16 grados usando los coeficientes encontrados
t_16 = 16
y_16 = a * t_16**3 + b * t_16**2 + c * t_16 + d

print(f"La vida media de la bacteria a 16 grados es: {y_16:.2f} minutos")
```

```
import numpy as np

# Definir las matrices A (coeficientes) y b (términos independientes)
A = np.array([[1, 1, 1],
              [30, 20, 10],
              [20, 0, 25]])

b = np.array([2.7, 56, 31])

# Resolver el sistema de ecuaciones para encontrar los precios A, B, C
precios = np.linalg.solve(A, b)

precio_A, precio_B, precio_C = precios
print(f"Precio de una unidad de A: {precio_A:.2f} €")
print(f"Precio de una unidad de B: {precio_B:.2f} €")
print(f"Precio de una unidad de C: {precio_C:.2f} €")


# (b) Encontrar el producto con el menor precio
menor_precio = min(precios)
producto_menor_precio = ["A", "B", "C"][np.argmin(precios)]
print(f"El producto con el menor precio es: {producto_menor_precio} con {menor_precio:.2f} €")

# (c) Encontrar el producto con el mayor precio
mayor_precio = max(precios)
producto_mayor_precio = ["A", "B", "C"][np.argmax(precios)]
print(f"El producto con el mayor precio es: {producto_mayor_precio} con {mayor_precio:.2f} €")


```
